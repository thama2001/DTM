### 2.2 Sri Lanka Weather Patterns (Tropical Monsoon Climate)

````python
def generate_sri_lanka_weather(timestamp, city):
    """
    Generate realistic weather conditions for Sri Lanka.
    Sri Lanka has tropical monsoon climate with two main monsoon seasons:
    - Southwest Monsoon (Yala): May-September (affects western/southern regions)
    - Northeast Monsoon (Maha): October-January (affects northern/eastern regions)
    """
    # Convert numpy.datetime64 to datetime for attribute access
    ts = pd.Timestamp(timestamp).to_pydatetime()
    month = ts.month
    hour = ts.hour
    
    # Regional classification
    western_southern = ['Colombo', 'Galle', 'Matara', 'Negombo', 'Katunayake']
    northern_eastern = ['Jaffna', 'Batticaloa', 'Trincomalee', 'Anuradhapura']
    hill_country = ['Kandy']
    
    # Base temperature (Sri Lanka average: 27-30Â°C coastal, 20-25Â°C hill country)
    if city in hill_country:
        base_temp = 22 + np.random.uniform(-2, 2)
    else:
        base_temp = 28 + np.random.uniform(-2, 3)
    
    # Daily temperature variation
    daily_var = 4 * np.sin((hour - 6) * np.pi / 12) if 6 <= hour <= 18 else -2
    temperature = base_temp + daily_var + np.random.normal(0, 1)
    
    # Humidity (generally high in Sri Lanka: 70-90%)
    humidity = np.random.uniform(70, 92)
    
    # Precipitation based on monsoon seasons
    if city in western_southern:
        # Southwest monsoon (May-Sept) affects these areas
        if month in [5, 6, 7, 8, 9]:
            precip_prob = 0.7
            precip_intensity = np.random.exponential(15)
        else:
            precip_prob = 0.3
            precip_intensity = np.random.exponential(5)
    elif city in northern_eastern:
        # Northeast monsoon (Oct-Jan) affects these areas
        if month in [10, 11, 12, 1]:
            precip_prob = 0.7
            precip_intensity = np.random.exponential(15)
        else:
            precip_prob = 0.25
            precip_intensity = np.random.exponential(4)
    else:  # Hill country - rain throughout year
        precip_prob = 0.5
        precip_intensity = np.random.exponential(10)
    
    precipitation = precip_intensity if np.random.random() < precip_prob else 0
    
    # Wind speed
    if month in [5, 6, 7, 8, 9] and city in western_southern:
        wind_speed = np.random.exponential(12) + 5  # Monsoon winds
    elif month in [10, 11, 12, 1] and city in northern_eastern:
        wind_speed = np.random.exponential(10) + 5
    else:
        wind_speed = np.random.exponential(5) + 2
    
    # Visibility (affected by rain and fog)
    if precipitation > 20:
        visibility = max(1, 10 - precipitation * 0.3 + np.random.normal(0, 1))
    elif hour in [5, 6, 7] and city in hill_country:
        visibility = np.random.uniform(2, 6)  # Morning mist in hills
    else:
        visibility = max(5, 10 + np.random.normal(0, 1))
    
    # Weather condition classification
    if precipitation > 25:
        condition = 'Heavy_Monsoon_Rain'
    elif precipitation > 10:
        condition = 'Moderate_Rain'
    elif precipitation > 2:
        condition = 'Light_Rain'
    elif visibility < 5:
        condition = 'Misty'
    elif wind_speed > 25:
        condition = 'Strong_Wind'
    elif humidity > 85 and temperature > 30:
        condition = 'Hot_Humid'
    else:
        condition = 'Clear'
    
    # Monsoon season indicator
    if month in [5, 6, 7, 8, 9]:
        monsoon_season = 'Southwest_Yala'
    elif month in [10, 11, 12, 1]:
        monsoon_season = 'Northeast_Maha'
    else:
        monsoon_season = 'Inter_Monsoon'
    
    return {
        'temperature_c': round(temperature, 1),
        'humidity_pct': round(humidity, 1),
        'precipitation_mm': round(max(0, precipitation), 2),
        'wind_speed_kmh': round(max(0, wind_speed), 1),
        'visibility_km': round(max(1, visibility), 1),
        'weather_condition': condition,
        'monsoon_season': monsoon_season
    }

# Test weather generation
print("\nğŸŒ§ï¸ Sri Lanka Weather Simulation Test:")
print("-"*60)
test_cities = ['Colombo', 'Kandy', 'Jaffna']
test_date = datetime(2024, 6, 15, 14, 0)  # During Southwest Monsoon

for city in test_cities:
    weather = generate_sri_lanka_weather(test_date, city)
    print(f"\nğŸ“ {city} (June - SW Monsoon):")
    for k, v in weather.items():
        print(f"   {k}: {v}")
````

````python
def calculate_congestion_sri_lanka(hour, day_of_week, weather, station_type, special_event=False):
    """
    Calculate traffic congestion for Sri Lankan transport conditions.
    Considers local factors like school hours, office hours, and festival periods.
    """
    # Sri Lanka specific rush hour patterns
    # Morning: 6:30-9:00 (School + Office)
    # Evening: 16:00-19:00 (School ends + Office)
    hourly_pattern = {
        0: 8, 1: 5, 2: 5, 3: 5, 4: 10, 5: 20, 6: 45, 7: 80, 8: 90, 9: 70,
        10: 50, 11: 55, 12: 60, 13: 55, 14: 50, 15: 60, 16: 75, 17: 92, 18: 85,
        19: 60, 20: 45, 21: 30, 22: 20, 23: 12
    }
    
    base_congestion = hourly_pattern[hour]
    
    # Day of week adjustment (Sunday is less busy, Saturday moderate)
    # In Sri Lanka, Poya days are holidays - simulated randomly
    day_factor = {
        0: 1.0, 1: 1.05, 2: 1.0, 3: 1.0, 4: 1.1, 5: 1.15,  # Mon-Fri increasing
        6: 0.7  # Sunday much quieter
    }
    
    congestion = base_congestion * day_factor[day_of_week]
    
    # Station type impact
    station_factor = {
        'Major_Terminal': 1.3,
        'Railway_Junction': 1.2,
        'Bus_Terminal': 1.25,
        'Airport': 0.9,
        'Coastal_Terminal': 1.0,
        'Northern_Terminal': 0.95,
        'Southern_Terminal': 1.0,
        'Heritage_City': 1.1,
        'Eastern_Terminal': 0.9,
        'Port_City': 1.05,
        'Airport_Connector': 1.0
    }
    
    congestion *= station_factor.get(station_type, 1.0)
    
    # Weather impact (monsoon rain significantly affects traffic)
    weather_impact = {
        'Clear': 0,
        'Hot_Humid': 5,
        'Strong_Wind': 8,
        'Misty': 12,
        'Light_Rain': 15,
        'Moderate_Rain': 25,
        'Heavy_Monsoon_Rain': 40  # Monsoon rain causes severe congestion
    }
    
    congestion += weather_impact.get(weather['weather_condition'], 0)
    
    # Heavy rain + rush hour = gridlock
    if weather['precipitation_mm'] > 15 and hour in [7, 8, 17, 18]:
        congestion += 20
    
    # Visibility impact
    if weather['visibility_km'] < 3:
        congestion += 15
    
    # Special events (festivals, cricket matches, etc.)
    if special_event:
        congestion += np.random.uniform(20, 40)
    
    # Add randomness
    congestion += np.random.normal(0, 5)
    
    # Bound to 0-100
    congestion = max(0, min(100, congestion))
    
    # Categorical level
    if congestion < 20:
        level = 'Free_Flow'
    elif congestion < 40:
        level = 'Light'
    elif congestion < 60:
        level = 'Moderate'
    elif congestion < 80:
        level = 'Heavy'
    else:
        level = 'Severe_Gridlock'
    
    return round(congestion, 2), level

print("âœ… Sri Lanka congestion calculation function defined")
````

````python
def generate_transport_record_sri_lanka(timestamp, mode, origin, destination):
    """
    Generate a comprehensive transport record for Sri Lanka network.
    """
    origin_config = STATIONS[origin]
    dest_config = STATIONS[destination]
    
    # Get weather for origin city
    weather = generate_sri_lanka_weather(timestamp, origin_config['city'])
    
    hour = timestamp.hour
    day_of_week = timestamp.weekday()
    month = timestamp.month
    
    # Special events (Poya days, festivals - about 8% of days)
    special_event = np.random.random() < 0.08
    
    # Congestion calculation
    congestion_index, congestion_level = calculate_congestion_sri_lanka(
        hour, day_of_week, weather, origin_config['hub_type'], special_event
    )
    
    # Mode-specific calculations
    mode_config = TRANSPORT_MODES[mode]
    
    # Calculate distance (based on coordinates)
    distance = np.sqrt(
        (origin_config['lat'] - dest_config['lat'])**2 +
        (origin_config['lon'] - dest_config['lon'])**2
    ) * 111  # Approximate km conversion
    
    # Base travel time
    base_time = (distance / mode_config['avg_speed']) * 60  # minutes
    
    # Adjusted travel time based on congestion and weather
    congestion_factor = 1 + (congestion_index / 100) * 0.6  # Sri Lanka has more congestion impact
    
    # Weather factor (monsoon rain significantly slows traffic)
    weather_factor = 1.0
    if weather['weather_condition'] == 'Heavy_Monsoon_Rain':
        weather_factor = 1.5
    elif weather['weather_condition'] == 'Moderate_Rain':
        weather_factor = 1.25
    elif weather['precipitation_mm'] > 0:
        weather_factor = 1.1
    
    # Mode-specific adjustments
    if mode == 'Three_Wheeler':
        # Tuk-tuks are very affected by rain and congestion
        weather_factor *= (1 + weather['precipitation_mm'] * 0.02)
        congestion_factor *= 0.85  # Can navigate through traffic better
    elif mode == 'Train':
        # Trains less affected by road congestion
        congestion_factor = 1 + (congestion_index / 100) * 0.15
    elif mode == 'Ferry':
        # Ferries affected by wind and waves
        if weather['wind_speed_kmh'] > 20:
            weather_factor *= 1.3
    
    actual_travel_time = base_time * congestion_factor * weather_factor + np.random.normal(0, 3)
    actual_travel_time = max(base_time * 0.8, actual_travel_time)
    
    # Passenger demand modeling
    base_demand = mode_config['base_capacity'] * 0.55
    footfall_factor = origin_config['daily_footfall'] / 100000
    
    # Time-based demand
    if hour in [7, 8, 9, 17, 18, 19]:
        demand_multiplier = 1.6 * footfall_factor
    elif hour in [0, 1, 2, 3, 4, 5]:
        demand_multiplier = 0.15 * footfall_factor
    else:
        demand_multiplier = 0.75 * footfall_factor
    
    # Weather impact on demand
    if weather['weather_condition'] in ['Heavy_Monsoon_Rain', 'Moderate_Rain']:
        if mode in ['Three_Wheeler', 'Taxi']:
            demand_multiplier *= 1.5  # More people want covered transport
        elif mode in ['SLTB_Bus', 'Private_Bus', 'Train']:
            demand_multiplier *= 1.3
    
    passenger_count = int(base_demand * demand_multiplier * np.random.uniform(0.7, 1.3))
    passenger_count = max(0, min(mode_config['base_capacity'], passenger_count))
    
    # Occupancy rate
    occupancy_rate = (passenger_count / mode_config['base_capacity']) * 100
    
    # Service reliability
    base_reliability = {'Train': 85, 'SLTB_Bus': 78, 'Private_Bus': 82, 
                        'Three_Wheeler': 92, 'Taxi': 94, 'Ferry': 80}
    reliability = base_reliability.get(mode, 85)
    reliability -= congestion_index * 0.15 + weather['precipitation_mm'] * 0.2
    reliability = max(50, min(100, reliability + np.random.normal(0, 3)))
    
    # Delay
    if reliability < 80:
        delay = np.random.exponential(6) * (100 - reliability) / 50
    else:
        delay = max(0, np.random.normal(0, 3))
    
    # Fuel consumption (liters per km)
    fuel_base = {'SLTB_Bus': 0.35, 'Private_Bus': 0.32, 'Train': 0.08, 
                 'Three_Wheeler': 0.05, 'Taxi': 0.08, 'Ferry': 2.0}
    fuel_consumption = fuel_base.get(mode, 0.1) * distance * (1 + congestion_index / 300)
    
    # CO2 emissions (kg)
    co2_factor = {'SLTB_Bus': 0.08, 'Private_Bus': 0.085, 'Train': 0.04,
                  'Three_Wheeler': 0.06, 'Taxi': 0.12, 'Ferry': 0.15}
    co2_emissions = co2_factor.get(mode, 0.1) * distance * (passenger_count + 1) / 10
    
    # Fare calculation (LKR - Sri Lankan Rupees)
    fare_per_km = {'SLTB_Bus': 2.5, 'Private_Bus': 3.0, 'Train': 2.0,
                   'Three_Wheeler': 50, 'Taxi': 80, 'Ferry': 15}
    fare_lkr = fare_per_km.get(mode, 5) * distance
    if mode in ['Three_Wheeler', 'Taxi']:
        fare_lkr += 100  # Base fare
    
    return {
        'timestamp': timestamp,
        'date': timestamp.date(),
        'hour': hour,
        'day_of_week': day_of_week,
        'day_name': timestamp.strftime('%A'),
        'month': month,
        'is_weekend': day_of_week >= 5,
        'is_rush_hour': hour in [7, 8, 9, 17, 18, 19],
        'transport_mode': mode,
        'origin_station': origin,
        'origin_city': origin_config['city'],
        'destination_station': destination,
        'destination_city': dest_config['city'],
        'origin_lat': origin_config['lat'],
        'origin_lon': origin_config['lon'],
        'dest_lat': dest_config['lat'],
        'dest_lon': dest_config['lon'],
        'hub_type': origin_config['hub_type'],
        'distance_km': round(distance, 2),
        'base_travel_time_mins': round(base_time, 2),
        'actual_travel_time_mins': round(actual_travel_time, 2),
        'delay_mins': round(delay, 2),
        'passenger_count': passenger_count,
        'vehicle_capacity': mode_config['base_capacity'],
        'occupancy_rate': round(occupancy_rate, 2),
        'congestion_index': congestion_index,
        'congestion_level': congestion_level,
        'temperature_c': weather['temperature_c'],
        'humidity_pct': weather['humidity_pct'],
        'precipitation_mm': weather['precipitation_mm'],
        'wind_speed_kmh': weather['wind_speed_kmh'],
        'visibility_km': weather['visibility_km'],
        'weather_condition': weather['weather_condition'],
        'monsoon_season': weather['monsoon_season'],
        'service_reliability': round(reliability, 2),
        'fuel_consumption_l': round(fuel_consumption, 3),
        'co2_emissions_kg': round(co2_emissions, 4),
        'fare_lkr': round(fare_lkr, 2),
        'special_event': special_event,
        'route_id': f"{mode[:3].upper()}_{origin[:3]}_{destination[:3]}_{random.randint(100, 999)}"
    }

print("âœ… Sri Lanka transport record generator defined")
````

### 3.4 ğŸ”— HYBRID MODEL: Weighted Ensemble of Random Forest + Gradient Boosting

The **Hybrid Model** combines predictions from both algorithms using a weighted average approach. This leverages the strengths of both methods:

- **Random Forest**: Good generalization, robust to noise
- **Gradient Boosting**: High accuracy on training patterns

**Hybrid Formula:**
$$\hat{y}_{hybrid} = w_1 \cdot \hat{y}_{RF} + w_2 \cdot \hat{y}_{GB}$$

where $w_1 + w_2 = 1$

````python
print("\n" + "="*80)
print("ğŸ”— HYBRID MODEL: RANDOM FOREST + GRADIENT BOOSTING ENSEMBLE")
print("="*80)

print("""
â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
â”‚                    HYBRID MODEL ARCHITECTURE                     â”‚
â”œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”¤
â”‚                                                                  â”‚
â”‚   Input Features (21 features)                                  â”‚
â”‚         â”‚                                                        â”‚
â”‚         â–¼                                                        â”‚
â”‚   â”Œâ”€â”€â”€â”€â”€â”´â”€â”€â”€â”€â”€â”                                                 â”‚
â”‚   â”‚           â”‚                                                  â”‚
â”‚   â–¼           â–¼                                                  â”‚
â”‚ â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â” â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”                                     â”‚
â”‚ â”‚  RANDOM   â”‚ â”‚  GRADIENT â”‚                                     â”‚
â”‚ â”‚  FOREST   â”‚ â”‚  BOOSTING â”‚                                     â”‚
â”‚ â”‚ (200 trees)â”‚ â”‚ (200 trees)â”‚                                    â”‚
â”‚ â””â”€â”€â”€â”€â”€â”¬â”€â”€â”€â”€â”€â”˜ â””â”€â”€â”€â”€â”€â”¬â”€â”€â”€â”€â”€â”˜                                     â”‚
â”‚       â”‚             â”‚                                            â”‚
â”‚       â–¼             â–¼                                            â”‚
â”‚    Pred_RF       Pred_GB                                        â”‚
â”‚       â”‚             â”‚                                            â”‚
â”‚       â””â”€â”€â”€â”€â”€â”€â”¬â”€â”€â”€â”€â”€â”€â”˜                                           â”‚
â”‚              â–¼                                                   â”‚
â”‚    â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”                                          â”‚
â”‚    â”‚ WEIGHTED AVERAGE â”‚                                          â”‚
â”‚    â”‚ w1=0.45, w2=0.55 â”‚                                          â”‚
â”‚    â””â”€â”€â”€â”€â”€â”€â”€â”€â”¬â”€â”€â”€â”€â”€â”€â”€â”€â”˜                                          â”‚
â”‚             â–¼                                                    â”‚
â”‚    â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”                                          â”‚
â”‚    â”‚ HYBRID PREDICTIONâ”‚                                          â”‚
â”‚    â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜                                          â”‚
â”‚                                                                  â”‚
â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜
""")

# Find optimal weights using grid search
print("\nâ³ Finding optimal ensemble weights...")

best_r2 = 0
best_weights = (0.5, 0.5)

weight_results = []
for w1 in np.arange(0.1, 1.0, 0.05):
    w2 = 1 - w1
    y_pred_hybrid = w1 * y_pred_rf_reg + w2 * y_pred_gb_reg
    r2 = r2_score(y_test_reg, y_pred_hybrid)
    weight_results.append({'w_RF': w1, 'w_GB': w2, 'R2': r2})
    if r2 > best_r2:
        best_r2 = r2
        best_weights = (w1, w2)

print(f"\nâœ… Optimal Weights Found:")
print(f"   â€¢ Random Forest Weight (w1): {best_weights[0]:.2f}")
print(f"   â€¢ Gradient Boosting Weight (w2): {best_weights[1]:.2f}")
````

````python
# Create Hybrid Model Predictions
print("\n" + "-"*80)
print("ğŸ”„ Generating Hybrid Model Predictions...")

# Regression: Weighted average
y_pred_hybrid_reg = best_weights[0] * y_pred_rf_reg + best_weights[1] * y_pred_gb_reg

# Classification: Weighted voting (soft voting)
y_pred_hybrid_proba = best_weights[0] * y_pred_rf_proba + best_weights[1] * y_pred_gb_proba
y_pred_hybrid_cls = np.argmax(y_pred_hybrid_proba, axis=1)

# Calculate Hybrid Metrics - Regression
hybrid_mse = mean_squared_error(y_test_reg, y_pred_hybrid_reg)
hybrid_rmse = np.sqrt(hybrid_mse)
hybrid_mae = mean_absolute_error(y_test_reg, y_pred_hybrid_reg)
hybrid_r2 = r2_score(y_test_reg, y_pred_hybrid_reg)

# Calculate Hybrid Metrics - Classification
hybrid_accuracy = accuracy_score(y_test_cls, y_pred_hybrid_cls)
hybrid_precision = precision_score(y_test_cls, y_pred_hybrid_cls, average='weighted')
hybrid_recall = recall_score(y_test_cls, y_pred_hybrid_cls, average='weighted')
hybrid_f1 = f1_score(y_test_cls, y_pred_hybrid_cls, average='weighted')

print("\n" + "="*80)
print("ğŸ“Š HYBRID MODEL REGRESSION RESULTS")
print("="*80)
print(f"\n   ğŸ¯ RÂ² Score:           {hybrid_r2:.4f} ({hybrid_r2*100:.2f}%)")
print(f"   ğŸ“‰ RMSE:               {hybrid_rmse:.4f}")
print(f"   ğŸ“‰ MAE:                {hybrid_mae:.4f}")
print(f"   ğŸ“‰ MSE:                {hybrid_mse:.4f}")

print("\n" + "="*80)
print("ğŸ“Š HYBRID MODEL CLASSIFICATION RESULTS")
print("="*80)
print(f"\n   ğŸ¯ Accuracy:           {hybrid_accuracy:.4f} ({hybrid_accuracy*100:.2f}%)")
print(f"   ğŸ¯ Precision:          {hybrid_precision:.4f} ({hybrid_precision*100:.2f}%)")
print(f"   ğŸ¯ Recall:             {hybrid_recall:.4f} ({hybrid_recall*100:.2f}%)")
print(f"   ğŸ¯ F1-Score:           {hybrid_f1:.4f} ({hybrid_f1*100:.2f}%)")

hybrid_results = {
    'Algorithm': 'Hybrid (RF + GB)',
    'R2_Score': hybrid_r2,
    'RMSE': hybrid_rmse,
    'MAE': hybrid_mae,
    'MSE': hybrid_mse
}

hybrid_cls_results = {
    'Algorithm': 'Hybrid (RF + GB)',
    'Accuracy': hybrid_accuracy,
    'Precision': hybrid_precision,
    'Recall': hybrid_recall,
    'F1_Score': hybrid_f1
}
````

````python
# Weather-Driven Route Optimization Class

class WeatherDrivenRouteOptimizer:
    """
    Weather-Driven Route Optimization for Sri Lanka Transport Network.
    Uses the trained Hybrid ML Model for predictions.
    """
    
    def __init__(self, rf_model, gb_model, weights, scaler, le_mode, le_weather, le_monsoon, le_hub):
        self.rf_model = rf_model
        self.gb_model = gb_model
        self.weights = weights
        self.scaler = scaler
        self.le_mode = le_mode
        self.le_weather = le_weather
        self.le_monsoon = le_monsoon
        self.le_hub = le_hub
        
        # Weather impact factors for Sri Lanka
        self.weather_penalties = {
            'Clear': 1.0,
            'Hot_Humid': 1.05,
            'Strong_Wind': 1.15,
            'Misty': 1.2,
            'Light_Rain': 1.25,
            'Moderate_Rain': 1.4,
            'Heavy_Monsoon_Rain': 1.7
        }
        
        self.mode_weather_sensitivity = {
            'SLTB_Bus': {'rain': 0.3, 'wind': 0.1},
            'Private_Bus': {'rain': 0.3, 'wind': 0.1},
            'Train': {'rain': 0.1, 'wind': 0.05},
            'Three_Wheeler': {'rain': 0.6, 'wind': 0.3},
            'Taxi': {'rain': 0.2, 'wind': 0.1},
            'Ferry': {'rain': 0.4, 'wind': 0.5}
        }
    
    def predict_congestion(self, features):
        """
        Predict congestion using the Hybrid Model.
        """
        rf_pred = self.rf_model.predict(features)
        gb_pred = self.gb_model.predict(features)
        hybrid_pred = self.weights[0] * rf_pred + self.weights[1] * gb_pred
        return hybrid_pred
    
    def get_mode_recommendations(self, weather_condition, precipitation, wind_speed):
        """
        Get transport mode recommendations based on weather.
        """
        recommendations = []
        
        for mode, sensitivity in self.mode_weather_sensitivity.items():
            # Calculate suitability score
            rain_penalty = 1 - (precipitation * sensitivity['rain'] / 50)
            wind_penalty = 1 - (max(0, wind_speed - 10) * sensitivity['wind'] / 30)
            
            base_score = self.weather_penalties.get(weather_condition, 1.0)
            suitability = max(0, min(100, (rain_penalty * wind_penalty * 100) / base_score))
            
            if weather_condition == 'Heavy_Monsoon_Rain' and mode == 'Three_Wheeler':
                suitability *= 0.3
            elif weather_condition == 'Strong_Wind' and mode == 'Ferry':
                suitability *= 0.4
            
            if suitability >= 80:
                rec = 'Highly Recommended'
            elif suitability >= 60:
                rec = 'Recommended'
            elif suitability >= 40:
                rec = 'Use with Caution'
            else:
                rec = 'Not Recommended'
            
            recommendations.append({
                'Mode': mode,
                'Suitability_Score': round(suitability, 1),
                'Recommendation': rec
            })
        
        return sorted(recommendations, key=lambda x: x['Suitability_Score'], reverse=True)

# Initialize optimizer
optimizer = WeatherDrivenRouteOptimizer(
    rf_regressor, gb_regressor, best_weights, scaler,
    le_mode, le_weather, le_monsoon, le_hub
)

print("\nâœ… Weather-Driven Route Optimizer initialized with Hybrid Model")
````
